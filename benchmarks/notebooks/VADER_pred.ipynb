```python
df_temp = pd.read_parquet(DATA_DIR + list_transcript_name[0])
df_temp = df_temp.sort_values(by=['transcriptid', 'transcriptcomponentid'])
```

```python
analyzer = SentimentIntensityAnalyzer()

def get_vader_sentiment(text):
    try:
        return analyzer.polarity_scores(text)['compound']  # Returns overall sentiment score
    except:
        return np.nan
```

```python
df_temp['sentiment_score'] = df_temp['componenttext'].progress_apply(get_vader_sentiment)
```

```python
df_temp_save = df_temp[['transcriptcomponentid', 'transcriptid', 'componentorder', 'transcriptcomponenttypeid', 'transcriptpersonid', 'sentiment_score']].copy()
```

```python
df_score_raw = df_temp_save[['transcriptcomponentid', 'transcriptid', 'componentorder', 'transcriptcomponenttypeid', 'transcriptpersonid', 'sentiment_score']].copy().reset_index(drop=True)

file_name = list_transcript_name[0].split('.')[0]
df_score_raw.to_parquet(DATA_DIR + f"sentiment_score_{file_name}.parquet")
```

# compute scores

```python
# for file_name in tqdm(list_transcript_name[1:]):
for file_name in tqdm(list_transcript_name[3:]):
    print(file_name)
    df_temp = pd.read_parquet(DATA_DIR + file_name)
    df_temp = df_temp.sort_values(by=['transcriptid', 'transcriptcomponentid'])
    df_temp['sentiment_score'] = df_temp['componenttext'].progress_apply(get_vader_sentiment)

    # save result
    df_temp_save = df_temp[['transcriptcomponentid', 'transcriptid', 'componentorder', 'transcriptcomponenttypeid', 'transcriptpersonid', 'sentiment_score']].copy()
    file_name = file_name.split('.')[0]
    df_temp_save.to_parquet(DATA_DIR + f"sentiment_score_{file_name}.parquet")
    print(DATA_DIR + f"sentiment_score_{file_name}.parquet")
```
