```python
from google.colab import drive
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
drive.mount('/content/drive')
DATA_DIR = "/content/drive/My Drive/data/"
## process your model_result
metrics_result = pd.read_csv(DATA_DIR + "metrics.csv")[["transcriptid","label"]]
model_result = pd.read_csv(DATA_DIR + "deepseek_chat_results_all.csv").rename(columns={"transcript_id":"transcriptid"})
model_result["sentiment"] = model_result["sentiment"].map({"neutral":0, "positive":1, "negative":-1})
## your model_result should have two columns "transcriptid" and "sentiment". for sentiment, it should only have 1,0,-1
model_result = pd.merge(model_result, metrics_result, on="transcriptid")
print("acc: ", accuracy_score(model_result["sentiment"], model_result["label"]))
print("precision: ", precision_score(model_result["sentiment"], model_result["label"], average= "weighted"))
print("recall: ", recall_score(model_result["sentiment"], model_result["label"], average= "weighted"))
print("f1: ", f1_score(model_result["sentiment"], model_result["label"], average= "weighted"))
# print("confusion matrix: \n", confusion_matrix(model_result["sentiment"], model_result["label"]))
```
