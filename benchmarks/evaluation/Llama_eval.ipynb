```python
from google.colab import drive
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
drive.mount('/content/drive')
DATA_DIR = "/content/drive/My Drive/DL Proj/data/"
```

```python
# Load the datasets
metrics_result = pd.read_csv(DATA_DIR + "metrics.csv")[["transcriptid", "label"]]
model_result = pd.read_csv(DATA_DIR + "llama_3_2_results.csv").rename(columns={"transcript_id": "transcriptid"})

# Map sentiment values to numbers
model_result["sentiment"] = model_result["sentiment"].map({"neutral": 0, "positive": 1, "negative": -1})

# Merge the datasets
model_result = pd.merge(model_result, metrics_result, on="transcriptid")

# Option 1: Drop rows with NaN values in either sentiment or label columns
clean_result = model_result.dropna(subset=['sentiment', 'label'])
print(f"Number of samples after removing NaN values: {len(clean_result)} out of {len(model_result)}")
print(f"Accuracy score: {accuracy_score(clean_result['sentiment'], clean_result['label'])}")
print("precision: ", precision_score(clean_result["sentiment"], clean_result["label"], average= "weighted"))
print("recall: ", recall_score(clean_result["sentiment"], clean_result["label"], average= "weighted"))
print("f1: ", f1_score(clean_result["sentiment"], clean_result["label"], average= "weighted"))

# Option 2: Fill NaN values with a default value (e.g., 0 for neutral sentiment)
# model_result['sentiment'] = model_result['sentiment'].fillna(0)
# model_result['label'] = model_result['label'].fillna(0)
# print(f"Accuracy score (after filling NaNs with 0): {accuracy_score(model_result['sentiment'], model_result['label'])}")

# Option 3: Check where the NaN values are coming from
# print("NaN counts in sentiment column:", model_result['sentiment'].isna().sum())
# print("NaN counts in label column:", model_result['label'].isna().sum())
```
